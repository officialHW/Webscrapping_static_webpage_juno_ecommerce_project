```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from sqlalchemy import create_engine
```

## Data Extraction

```python
# firstly initiate

url = "https://www.jumia.com.ng/mlp-working-from-anywhere/laptops/hp/"

r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
```

```python
print(r)
```

```python
# now I want to get the laptop Product names

product_names = [] #Here creating an empty list which i want the names of the products in that web

names = soup.find_all("h3", class_ = "name") #Here inside the names variable, i am requesting for all the product names that are present on that page, as seen in the inspect that the tag in the website for the prices is <h3 class="name">

for name in names: #Then, for all the product names present , i will extract them all as a string
    
    product_names.append(name.string) #And put them inside the empty list called product_names which i created
```

```python

product_names #As seen, the result shows all the laptop product names on the website page
```

```python
# Now I will check the product current price

sales_price = [] # here creating an empty list which i want the prices in webpage

prices = soup.find_all("div", class_ = "prc") # here inside the prices variable, i am requesting for all the prices that are present on that page, as seen in the inspect that the tag in the website for the prices is <div class="prc">

for prc in prices: # then, for all the price present , i will extract them all as a string

    sales_price.append(prc.string) #  and put them inside the empty list called current_price which i created
```

```python
sales_price
```

```python
# former price (using the same logic that was used in the product_names and the current_price)
original_price = []
old_prices = soup.find_all('div', class_="old")
for old in old_prices:
    original_price.append(old.string)
original_price
```

```python
# discount percentages (using the same logic that was used in the previouse)
discount_percs = []
discounts = soup.find_all('div', class_="bdg _dsct _sm")
for discount in discounts:
    discount_percs.append(discount.string)
discount_percs
```

##### putting the data into a data frame

```python
# but first creating it as a dictionary
data = {
    'title': product_names,
    'sales_price': sales_price,
    'original_price': original_price,
    'discount': discount_percs
    }
# now putting it into a datafram and noting that its from the dictionary
laptop_df = pd.DataFrame.from_dict(data, orient = 'index')
laptop_df =laptop_df.transpose() #to invert the columns.
laptop_df

```

#### Now we consider that jumia does not have just 1 page, and we need the result for all the pages,
##### now what we will do is to update the url and scrape it so it can show result for all the pages

```python
# extraction of all pages

pages = [x for x in range(1,51)]

# rewriting the code
product_names = []
sales_price = []
original_price = []
discount_percs = []

# updating the url
for page in pages:# firstly initiate
    url = f"https://www.jumia.com.ng/mlp-working-from-anywhere/laptops/hp/?page={page}#catalog-listing"

    r = requests.get(url)

    soup = BeautifulSoup(r.text, "lxml")
    
# pasting remaining codes
    # Product Names
    products = soup.find_all('h3', class_ = "name")
    for name in names:
        product_names.append(name.text)
    
    #Sales Price
    prices = soup.find_all('div', class_ = "prc")
    for prc in prices:
        sales_price.append(prc.text)
    
    # Old price
    old_prices = soup.find_all('div', class_="old")
    for old in old_prices:
        original_price.append(old.text)
    
        
    # Discount
    percent = soup.find_all('div', class_="bdg _dsct _sm")
    for per in percent:
        discount_percs.append(discount.text)

# putting all into a dataframe for proper presentation
data = {
    'title': product_names,
    'sales_price': sales_price,
    'original_price': original_price,
    'discount': discount_percs
    }

laptop_df = pd.DataFrame.from_dict(data, orient = 'index')
laptop_df =laptop_df.transpose() #to invert the columns.
laptop_df
```
